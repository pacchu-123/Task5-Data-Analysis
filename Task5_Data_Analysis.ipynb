```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales.csv")
df.head()
```

```python
df.info()
df.describe()
df.shape
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
product_sales
```

```python
product_sales.plot(kind="bar")
plt.title("Sales by Product")
plt.show()
```
